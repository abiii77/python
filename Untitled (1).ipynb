```python
import pandas as pd

# Load the dataset
file_path = 'C:\\Users\\Lenovo\\Desktop\\PS_20174392719_1491204439457_log.csv'
data = pd.read_csv(file_path)

# Display the first few rows of the dataset
print(data.head())

# Display basic information about the dataset
print(data.info())

# Display summary statistics
print(data.describe())
```

```python
# Check for missing values
missing_values = data.isnull().sum()
print("Missing values:\n", missing_values)

# Drop rows with missing values (if any)
data_cleaned = data.dropna()

# Check for duplicates
duplicates = data_cleaned.duplicated().sum()
print("Duplicates:\n", duplicates)

# Remove duplicates
data_cleaned = data_cleaned.drop_duplicates()

# Ensure all data types are appropriate
print(data_cleaned.dtypes)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Summary statistics
summary_statistics = data_cleaned.describe()
print(summary_statistics)

# Distribution analysis - Histograms
data_cleaned.hist(bins=50, figsize=(20, 15))
plt.show()

# Correlation analysis - Exclude non-numeric columns
numeric_columns = data_cleaned.select_dtypes(include=['number']).columns
correlation_matrix = data_cleaned[numeric_columns].corr()
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.show()
```

```python
# Box plots for numerical features
numerical_features = ['amount', 'oldbalanceOrg', 'newbalanceOrig', 'oldbalanceDest', 'newbalanceDest']
for feature in numerical_features:
    plt.figure(figsize=(10, 6))
    sns.boxplot(x=data_cleaned[feature])
    plt.title(f'Box plot of {feature}')
    plt.show()

# Scatter plots for selected pairs of variables
sns.pairplot(data_cleaned[numerical_features])
plt.show()
```

```python
from scipy import stats

# Hypothesis testing - Compare means of fraudulent vs non-fraudulent transactions
fraud = data_cleaned[data_cleaned['isFraud'] == 1]['amount']
non_fraud = data_cleaned[data_cleaned['isFraud'] == 0]['amount']

t_stat, p_value = stats.ttest_ind(fraud, non_fraud)
print(f"T-statistic: {t_stat}, P-value: {p_value}")
```
